### Python Packages for Data Science

  - **Pandas** (Estructuras de datos y herramientas)

  - **SciPy** (Integrales, ecuaciones diferenciales, optimización)

### Importing and Exporting Data in Python

Importar Datos: Es el proceso de carga y lectura de datos en Python de varias fuentes o recursos.

- Formato: CSV, JSON, XLSX, HDF, ...
- Path: (Dirección de la ubicación de la fuente de datos)

```python
import ssl

try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    # Legacy Python that doesn't verify HTTPS certificates by default
    pass
else:
    # Handle target environment that doesn't support HTTPS verification
    ssl._create_default_https_context = _create_unverified_https_context

# Importando datos
import pandas as pd

# Definicion de variable con la dirección de los datos
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

# Leer los daots de la ubicacion; los datos requeiren de un header al usar esta función "read_csv"
df = pd.read_csv(url, header = None)

# Muestra los primeros n registros de los datos
df.head(5)

# Muestra los ultimos n registros de los datos 
#print(df.tail(5))
```

```python
# Agregando headers al DataFrame
headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", "body-style", "drive-wheels", "engine-location", "wheel-base", "length", "width", "height", "curb-weight", "engine-type", "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]
df.columns = headers
df.head(5)
```

```python
# Exportar un Dataframe de pandas a un CSV
import os
path = str(os.getcwd()) + '/data/automobile.csv'
print(path)
df.to_csv(path)
```
